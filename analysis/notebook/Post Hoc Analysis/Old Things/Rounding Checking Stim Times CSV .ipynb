```python
import pandas as pd

# 1. Read the CSV file into a DataFrame
csv_file_path = "Checking Stim Times High.csv"  # Replace with the actual path to your CSV file
df = pd.read_csv(csv_file_path)

# 2. Round the values to two decimal places
df = df.round(2)

# 3. Save the updated DataFrame to a new CSV file
output_csv_file_path = "Rounded Stim Times.csv"  # Replace with your desired output file path
df.to_csv(output_csv_file_path, index=False)

# Display the first few rows of the updated DataFrame
df.head()
```
