```python
# Function to filter headers by numbers and type of values
def filter_headers(df):
    numeric_headers = []
    string_headers = []

    for column in df.columns:
        if df[column].dtype in ['int64', 'float64']:
            numeric_headers.append(column)
        else:
            string_headers.append(column)

    return numeric_headers, string_headers

# Input the path to the CSV file
csv_path = input("Enter the path to the CSV file: ")

# Check if the file exists
if not os.path.isfile(csv_path):
    print("File not found.")
    exit()

# Load the CSV file into a pandas DataFrame
df = pd.read_csv(csv_path, delimiter=';')

# Get the headers
headers = list(df.columns)

while True:
    # Print the headers with numbers for selection
    print("\nHeaders:")
    for i, header in enumerate(headers):
        print(f"{i + 1}. {header}")

    # Prompt the user to select a header
    selected_index = int(input("Enter the number corresponding to the header you want to analyze: ")) - 1
    selected_header = headers[selected_index]

    # Filter the DataFrame to show only the selected header
    selected_column = df[selected_header]

    # Display the selected column
    print(f"\nSelected column ({selected_header}):")
    print(selected_column)

    # Prompt the user to enter a value for filtering
    filter_value = input("Enter the value to filter the selected column: ")

    # Filter the DataFrame to keep only rows where the selected column matches the entered value
    filtered_df = df[df[selected_header] == filter_value]

    # Display the filtered DataFrame
    print("\nFiltered DataFrame:")
    print(filtered_df)

    # Ask if the user wants to continue filtering
    choice = input("Do you want to continue filtering? (yes/no): ")
    if choice.lower() != 'yes':
        break
```
