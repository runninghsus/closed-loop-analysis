## Adds Stim Count and Frame Rate to Array 

```python
for z in BehaviorTime:
    behavior = z[0]
    time= z[1]
behavior_type_and_duration = []
current_behavior = BehaviorTime[0][0]
current_time = BehaviorTime[0][1]  
duration = 0
stim_count=0
frames = 0

for behavior, time in BehaviorTime:
    if behavior == current_behavior:
        duration += 1
        if duration == 1:
            start_time= current_time
    else:
        if duration==1:
            end_time=start_time+1
        else:
            end_time= start_time + duration
        frames= duration*3
        stim_count= round(duration/3)
        behavior_type_and_duration.append((current_behavior, duration,frames,stim_count,start_time, end_time))
        current_behavior = behavior
        current_time = time
        start_time= time
        duration = 1
        end_time= time+1
        
# Append the final (current_behavior, current_time, count) tuple outside the loop
behavior_type_and_duration.append((current_behavior, duration, frames, stim_count, start_time, end_time))

# Make a copy of the original behavior_type_and_duration list
behaviortdt= list(behavior_type_and_duration)
```

## Add Bout Count to the Array 

```python
enumerated_behaviortdt = []
behavior_index = {}  # Dictionary to store behavior and its corresponding index

for behavior, duration, frames, stim_count, start_time, end_time in behaviortdt:
    if behavior not in behavior_index:
        behavior_index[behavior] = 1
    else:
        behavior_index[behavior] += 1
    
    bout = behavior_index[behavior]
    enumerated_behaviortdt.append((behavior, duration,frames, bout, stim_count,start_time,end_time))

print(enumerated_behaviortdt)
```

# Stim Count Calcultaion

```python
BCBDST = enumerated_behaviortdt
behavior_labels = {
    0.0: "Left Turns",
    1.0: "Right Turns",
    2.0: "Locomotion",
    3.0: "Face Groom",
    4.0: "All Other Groom",
    5.0: "All Other",
}

bdbt_dict = {}  # Dictionary to store lists of behaviors

for behavior, duration, frames, bout, stim_count, start_time,end_time in BCBDST:
    behavior_label = behavior_labels.get(behavior, "Unknown Behavior")
    if behavior_label not in bdbt_dict:
        bdbt_dict[behavior_label] = []
    bdbt_dict[behavior_label].append((behavior, stim_count, frames, bout, duration, start_time,end_time))
```

## Stim Count and Average Duration 

```python
from tabulate import tabulate
stimcount_by_key = {}

def round_float(value):
    return round(value, 3)

average_duration_30min_by_key = {}

for key, tuples_list in bdbt_dict.items():
    total_duration = 0  # Initialize the total duration to 0
    count = 0  # Initialize a count variable to track the number of valid tuples
    
    for tuple_item in tuples_list:
        if tuple_item[5] > 3000 and tuple_item[6] <= 21000:  # Check the conditions
            total_duration += tuple_item[4]  # Add the duration to the total
            count += 1  # Increment the count
            
    if count > 0:
        average_duration = round(((total_duration / count)/10), 2)  # Calculate average and round it
    else:
        average_duration = 0  # Set average to 0 if there are no valid tuples
    
    average_duration_30min_by_key[key] = average_duration

# Printing the results in a tabular format
table_data = [(key, avg_duration) for key, avg_duration in average_duration_30min_by_key.items()]
headers = ["Key", "Average Duration During 30 Min Stim Block (seconds)"]

from tabulate import tabulate

table = tabulate(table_data, headers, tablefmt="pretty")
print(table)


```

##  Number of Bouts

```python
stim_block_bout_count_by_key = {}

for key, tuples_list in bdbt_dict.items():
    bout_count = sum(1 for tuple_item in tuples_list if 3000 < tuple_item[5] and tuple_item[6] <= 21000)
    stim_block_bout_count_by_key[key] = bout_count

table_data = [(key, count) for key, count in stim_block_bout_count_by_key.items()]
headers = ["Key", "30 minute Stim Block Bout Count"]

table = tabulate(table_data, headers, tablefmt="pretty")
print(table)
```

```python
# Initialize a dictionary to store the products
products_by_key = {}

# Calculate the products and store them in the dictionary
for key in bdbt_dict.keys():
    if key in average_duration_30min_by_key and key in stim_block_bout_count_by_key:
        product = average_duration_30min_by_key[key] * stim_block_bout_count_by_key[key]
        products_by_key[key] = product

# Calculate the total time of the stim block
total_time_of_stim_block = round((sum(products_by_key.values()))/60,2)

# Print the products and the total time of the stim block
for key, product in products_by_key.items():
    print(f"Key: {key}, Product: {product}")

print(f"Total Time of the Stim Block: {total_time_of_stim_block} minutes")
```

```python
import os
import csv

# Define the output CSV file path
output_csv_file = 'Checking Stim Times High.csv'

# Define the headers as you specified
headers = [
    "Mouse",
    "Genotype",
    "Date",
    "Selected Behavior",
    "Fiber Connection",
    "Sham",
    "Total Time Spent In Stim (minutes)",
    "All Other",
    "Locomotion",
    "Left Turns",
    "Right Turns",
    "Face Groom", 
    "All Other Groom",# Include header keys from products_by_key
]


# Define the field names for the CSV
fieldnames = headers

# Check if the CSV file already exists
csv_file_exists = os.path.isfile(output_csv_file)

# Open the CSV file in append mode ('a') if it already exists, otherwise create it
with open(output_csv_file, 'a', newline='') as csvfile:
    writer = csv.writer(csvfile)

    # Write the header row only if the file is newly created
    if not csv_file_exists:
        writer.writerow(fieldnames)
    
    # Iterate through the input file paths
    for file_path in infiles:
        AO=products_by_key.get("All Other")
        Loco=products_by_key.get("Locomotion")
        LT=products_by_key.get("Left Turns")
        RT=products_by_key.get("Right Turns")
        FG=products_by_key.get("Face Groom")
        AOG=products_by_key.get("All Other Groom")
        sham="n"
        folders = os.path.dirname(file_path).split('/')
        if len(folders) >= 4:
            # Extract information from folder paths
            mouse_genotype = folders[2].split("-")
            date = folders[4].split("-")[0]

            selected_behavior = ""
            fiber_connection = ""

            # Check if the folder name contains "RT" or "rt" (Right Turn)
            folder_parts = folders[4].split("-")
            folder_1 = folder_parts[1].split(" ")
            
            if folder_1[0].lower() in ["rt", "rightturn"]:
                selected_behavior = "Right Turn"
                if folder_1[1].lower() in ["contra", "contralateral"]:
                    fiber_connection = "Contralateral"
                elif folder_1[1].lower() in ["ipsilateral", "ipsi"]:
                    fiber_connection = "Ipsilateral"
                else:
                    # Default value if none of the conditions are met
                    fiber_connection = "Bilateral"
                if len(folder_1) >= 3:
                    if folder_1[2] == "sham":
                        sham = "y"
                        
            elif folder_1[0].lower() == "face":
                selected_behavior = "Face Groom"
                fiber_connection = "Bilateral"
                if len(folder_1) >= 3:
                    if folder_1[2] == "sham":
                        sham = "y"
            else:
                selected_behavior = folder_1[0]
                fiber_connection = "Bilateral"
                if len(folder_1) >= 2:
                    if folder_1[1] == "sham":
                        sham = "y"

            # Create a row as a list of values
            row = [
                mouse_genotype[0],
                mouse_genotype[1],
                date,
                selected_behavior,
                fiber_connection,
                sham,
                total_time_of_stim_block,
                AO,
                Loco,
                LT,
                RT, 
                FG,
                AOG,
            ]

            # Write the row to the CSV file
            writer.writerow(row)
        else:
            print("Not enough elements in the folders list.")

# Print a message indicating the CSV file has been created or updated
if csv_file_exists:
    print(f"CSV file '{output_csv_file}' has been updated.")
else:
    print(f"CSV file '{output_csv_file}' has been created.")

print(folder_1)
```
